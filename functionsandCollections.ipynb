```python
# Functions

"""
1) built in
2) user defined
3) function arguments : 1) required arguments
                        2) default arguments
                        3) keyword arguments
                        4) variable length arguments (*args, **kwargs)
4) return statement


"""


def add(a, b): # user defined function
    return a + b
print(add(10, 20)) # function call with required arguments


def greet(name):
    print("Hello, " + name)
greet(name="Alice") # function call with required argument

def greet(name="Guest"): # function with default argument
    print("Hello, " + name)
greet() # function call without argument, uses default
greet("Bob") # function call with argument


def details(name,age): # keyword arguments
    print("Name:", name)
    print("Age:", age)
details(age=25, name="Charlie") # function call with keyword arguments

def add(*nums):
    return sum(nums)
print(add(1, 2, 3, 4, 5)) # function call with variable length arguments
```

### Collections

```python
## List

nums=[10,20,30]

nums.append(40) # add element to the end
nums.insert(1,15) # insert element at index 1
nums.remove(20) # remove element by value
nums.pop() # remove and return last element
nums.sort() # sort the list
nums.reverse() # reverse the list
nums.index(30) # get index of element
len(nums) # get length of the list
nums[1:3] # slicing
nums[::2] # slicing with step
# nums.clear() # clear the list
print(nums)
```

```python
## Tuple
# ordered, immutable collection, allows duplicate elements, faster than list

t=(1,2,3,2,3,4,2,4,3)
print(t[0])

print(t.count(2)) # count occurrences of an element
print(t.index(3)) # get index of an element
print(len(t)) # get length of the tuple
# t[0]=10 # error, tuples are immutable
print(t)

print(tuple(sorted(t))) # sort tuple

# slicing
print(t[1:5])

# tuple unpacking
t1=(10,20,30)
a,b,c=t1
print(a,b,c)

a1,b1,*c1=t
print(a1,b1,c1)
```
