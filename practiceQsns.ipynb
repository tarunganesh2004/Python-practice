```python
# Check for prime Number

def is_prime(n):
    if n<=1:
        return False
    
    # brute force
    for i in range(2,n):
        if n%i==0:
            return False
    return True

    # O(sqrt(n))
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True

n=89
if is_prime(n):
    print(f"{n} is a prime number")
else:
    print(f"{n} is not a prime number")
```

```python
# fibonacci series
def fibonacci_series(n):
    res=[]
    a,b=0,1
    for _ in range(n):
        c=a+b
        res.append(a)
        a=b
        b=c
    return res
n=10
print(fibonacci_series(n))

# recursion
def fib_rec(n):
    if n==0:
        return [0]
    if n==1:
        return [0,1]
    
    seq=fib_rec(n-1)
    seq.append(seq[-1]+seq[-2])
    return seq

print(fib_rec(10))

def fib_rec(n):
    if n == 0:
        return 0, [0]
    if n == 1:
        return 1, [0, 1]

    a, seq1 = fib_rec(n - 1)
    b, _ = fib_rec(n - 2)
    return a + b, seq1
```

```python
# sum of digits

# 1) using string conversion
def sum_digits(n):
    return sum(int(d) for d in str(n))

# using modulo
def sum_digits_mod(n):
    s=0
    while n>0:
        s+=n%10
        n=n//10
    return s

n=12345
print(sum_digits(n))
print(sum_digits_mod(n))
```

```python
# reverse a number
def reverse_number(n):
    rev=0
    while n>0:
        d=n%10
        rev=rev*10+d
        n=n//10
    return rev

n=12345
print(reverse_number(n))
```

```python
# print multiplication table
def multi(n):
    for i in range(1,11):
        print(f"{n} x {i} = {n*i}")
n=7
multi(n)
```

```python
# Find largest of 3 numbers using nested if
def largest(a,b,c):
    if a>=b:
        if a>=c:
            return a
        else:
            return c
    else:
        if b>=c:
            return b
        else:
            return c
n1,n2,n3=10,25,15
print(largest(n1,n2,n3))
```

```python
# count vowels in a string
def count_vowels(s):
    vowels="aeiouAEIOU"
    count=0
    for char in s:
        if char in vowels:
            count+=1
    return count

s="Hello World"
print(count_vowels(s))
```
