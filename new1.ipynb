```python
import re

s="1234215"

if re.fullmatch(r"[A-Za-z]+",s):
    print("only letters")
elif re.search(r"https?://(?:www\.)?[^/]+/([A-Za-z0-9_]+)",s):
    username=re.search(r"https?://(?:www\.)?[^/]+/([A-Za-z0-9_]+)",s).group(1)
    print("URL")
    print("username:",username)

elif re.fullmatch(r"[A-Z]+", s):
    print("Only capital letters")

elif re.fullmatch(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", s):
    print("Email")

elif re.fullmatch(r"\d{6,}", s):
    print("Digits (6 or more)")

else:
    print("No match")
```
