# File Handling


4Ô∏è‚É£ Opening a File (open())

Syntax
```python
file_object = open("filename", "mode")

Common File Modes (VERY IMPORTANT)

Mode	Meaning

r	Read (default)

w	Write (overwrite)

a	Append

x	Create new file

r+	Read + Write

w+	Write + Read

a+	Append + Read

rb+ Read + Write in binary(mostly preferred for positioing in file handling)

ab+ Append + Read in binary



-- to close a file --
file_object.close()

-- to read from a file --

content = file_object.read() -- reads entire file

line = file_object.readline() -- reads one line

lines = file_object.readlines() -- reads all lines into a list

-- to write to a file --

w mode destroys existing content

write() overwrites file 

file_object.write("text to write")

file_object.writelines(list_of_strings)

-- creating a file (using x mode)--

file_object = open("newfile.txt", "x")

file_object.write("This is a new file.")
file_object.close()
```

````python
# read() -- read full file
f=open("sample.txt","r")
content=f.read()
print(content)
f.close()
````

````python
# readlines() -- read all lines into a list
f=open("sample.txt","r")
print(f.readlines())
f.close()
````

````python
# writing to a file
f=open("data.txt","w") # w destroys old content(if the file exists) else creates a new file
f.write("Hello Python\n")
f.write("File Handling in Python\n")
f.close()
````

````python
# appending to a file (a mode )
f=open("data.txt","a") # a mode appends to the existing content
f.write("Appending new line 1\n")
f.write("Appending new line 2\n")
f.close()
````

````python
# create a file (x mode)
f=open("newfile.txt","x") # x mode creates a new file, if the file exists it raises an error
f.write("This is a new file created using x mode.\n")
f.close()
````

````python
# file pointer tell()and seek()
# tell() returns the current position of the file pointer
f=open("sample.txt","r")
print(f.tell())
print(f.read(10))  # read first 10 characters
print("Current file pointer position:",f.tell())

# seek() changes the file pointer position
f.seek(0)  # move pointer to the beginning
print(f.read(10))  # read first 10 characters again
f.close()

f=open("sample.txt","r")
f.seek(10) # move pointer to the 10th position and file starts reading from there
print(f.read(10))
f.close()
````

````python
# With statement for file handling(best practice)
with open("sample.txt","r") as f:
    content=f.read()
    # here l1 & ll are empty bcz f.read() takes the file pointer to end

    #so read again we should change the file pointer
    f.seek(0)
    l1=f.readline()
    ll=f.readlines()
    print(content)
    print(l1)
    print(ll)

    f.seek(10)
    print(f.read(10))
````

````python
# seek(offset,whence)
# whence=0 --> beginning of the file
# whence=1 --> current position
# whence=2 --> end of the file
with open("sample.txt","r") as f:
    print(f.tell()) # initially 0

    f.seek(6) # navigates to 12th position from beginning
    print(f.tell())

    # this doesnt work in r mode bcz Why Python forbids this üß†
# In text mode:

# Newlines may be translated (\r\n ‚Üî \n)

# Characters may be multi-byte (UTF-8)

# So Python cannot reliably move ‚Äú3 bytes forward‚Äù relative to current position.
    # f.seek(3,1) # navigates 3 positions ahead from current position
    print(f.tell())
    
# so we should open in binary mmode 
with open("sample.txt","rb+") as f:
    print(f.tell()) # initially 0

    f.seek(12) # navigates to 6th position from beginning
    print(f.tell())

    f.seek(3,1) # navigates 3 positions ahead from current position
    print(f.tell())

    f.seek(-3,2) # navigates 3 positions back from end of the file
    print(f.tell())
````

````python
# reading file line by line using for loop
with open("sample.txt","r") as f:
    for line in f:
        print(line)
````

````python
# check if file exists
import os
if os.path.exists("sample.txt"):
    print("File exists")

else:
    print("File does not exist")
````

````python
# file handling with exception handling
try:
    with open("test.txt") as f:
        print(f.read())
except FileNotFoundError:
    print("File not found. Please check the file path.")
````

````python
# writing multiple lines
lines=["Python\n","Java\n","C++\n","JavaScript\n"]
with open("data.txt","w") as f:
    f.writelines(lines) # replaces old content, to append use a
````

````python
# copying content from one file to another
with open("sample.txt","r") as source:
    with open("copy.txt","w") as dest:
        for line in source:
            dest.write(line)
````

````python
# count lines,words,characters in a file
with open("sample.txt","r") as f:
    text=f.read()
    lines=text.splitlines()
    words=text.split()
    characters=len(text)

print("Lines:",len(lines))
print("Words:",len(words))
print("Characters:",characters)
````

### Examples


````python
# print line numbers that are even
with open("data.txt","r") as f:
    for i,line in enumerate(f,start=1):
        if i%2==0:
            print(line.strip())
````

````python
# count number of vowels in a file
count=0
with open("sample.txt","r") as f:
    text=f.read()
    for ch in text:
        if ch in 'aeiouAEIOU':
            count+=1
print("Number of vowels:",count)
````

````python
# copy only lines containing a specific word from one file to another
word="Python"
with open("input.txt","r") as src,open("output.txt","w") as dest:
    for line in src:
        if word in line:
            dest.write(line)
````

````python
# reverse the entire file content
with open("data.txt","r") as f:
    lines=f.readlines()

with open("data.txt","w") as f:
    for line in reversed(lines):
        f.write(line)
````

````python
# merge two files into a third file
with open("file1.txt","r") as f1,open("file2.txt","r") as f2:
    content=f1.read()+"\n"+f2.read()
with open("merged.txt","w") as f:
    f.write(content)
````

````python
# remove duplicate lines from a file
s=set()
with open("data.txt" ,"r") as f:
    lines=f.readlines() # first read

with open("data.txt","w") as f:
    for line in lines:
        if line not in s:
            f.write(line)
            s.add(line)
````

````python
# copy  a file
with open("sample.txt","r") as src,open("sample_copy.txt","w") as dest:
    dest.write(src.read())
````

````python
# print longest line in a path
with open("data.txt", "r") as f:
    longest = max(f.readlines(), key=len)

print(longest.strip())
````
